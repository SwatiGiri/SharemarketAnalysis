```python
#importing dependencies 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing data
data=pd.read_csv("ETF composition - IRBO.csv",names=["Companies","ETF"])
data2=pd.read_csv("ETF composition - XWEB.csv",names=["Companies","ETF"])
```

```python
#Checking data
data.head()
```

```python
#checking second set of data
data2.head()
```

```python
#convert ETF in to float
type(data.ETF[0])
ETF_wo_per=[]
for i in data.ETF:
    x=float(i[:4])
    ETF_wo_per.append(x)
type(ETF_wo_per[0])
```

```python
data["ETF_wo_per"]=ETF_wo_per
```

```python
type(data2.ETF[0])
ETF_wo_per2=[]
for i in data2.ETF:
    x=float(i[:4])
    ETF_wo_per2.append(x)
type(ETF_wo_per2[0])
data2["ETF_wo_per"]=ETF_wo_per2
data2
```

```python
data.head()
```

```python
#finding common companies
common_companies=pd.core.common.intersection(data.Companies, data2.Companies).tolist()
```

```python
#list of common companies
common_companies
```

```python
#locating common companies in first dataset and storing it in dataframe
df=pd.DataFrame(data=data.loc[data["Companies"].isin(common_companies)])
df
```

```python
#finding percentage of ETF in common companies and uncommon companies
commoncompnies_ETF_df=sum(df.ETF_wo_per)
uncommoncompnies_ETF_df=sum(data.ETF_wo_per)-commoncompnies_ETF_df
uncommoncompnies_ETF_df
```

```python
#locating common companies in second dataset and storing it in dataframe
df2=pd.DataFrame(data=data2.loc[data["Companies"].isin(common_companies)])
df2
commoncompnies_ETF_df2=sum(df2.ETF_wo_per)
uncommoncompnies_ETF_df2=sum(data2.ETF_wo_per)-commoncompnies_ETF_df2
uncommoncompnies_ETF_df2
```

```python
df2
```

```python
#plotting all companies and their ETF percentage in first data
labels= data.Companies
sizes = data.ETF_wo_per
plt.pie(sizes, labels=labels,labeldistance=1.1,radius=5,autopct='%1.1f%%',center=(0, 0),rotatelabels=45)
plt.show()
```

```python
#plotting all companies and their ETF percentage in second data
labels= data2.Companies
sizes = data2.ETF_wo_per
plt.pie(sizes, labels=labels,labeldistance=1.1,radius=5,autopct='%1.1f%%',center=(0, 0),rotatelabels=45)
plt.show()
```

```python
# percentage ETF of common companies and uncommon companies in first dataset
labels= ["Common Companies", "Uncommon companies"]
sizes = [commoncompnies_ETF_df,uncommoncompnies_ETF_df]
plt.title("ETF difference of common companies and uncommon companies in dataset 1")
plt.pie(sizes, labels=labels,labeldistance=1.1,radius=2,autopct='%1.1f%%',center=(0, 0),explode = (0.1, 0))
plt.axis("equal")
plt.show()
```

```python
# percentage ETF of common companies and uncommon companies in second dataset
labels= ["Common Companies", "Uncommon companies"]
sizes = [commoncompnies_ETF_df2,uncommoncompnies_ETF_df2]
plt.title("ETF difference of common companies and uncommon companies in dataset 2")
plt.pie(sizes, labels=labels,labeldistance=1.1,radius=2,autopct='%1.1f%%',center=(0, 0),explode = (0.1, 0))
plt.axis("equal")
plt.show()
```
